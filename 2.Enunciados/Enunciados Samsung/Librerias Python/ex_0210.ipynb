### 1. Visualización con objetos Matplotlib:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### 1.1. Visualización con el objeto figure:

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
fig0=plt.figure()
axes0 = fig0.add_axes([0,0,1,1])             # Izquierda, abajo, ancho, alto.
axes0.plot(x,y,color='red',linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
plt.show()
```

```python
fig0=plt.figure()
axes0 = fig0.add_axes([0,0,1,1])             # Izquierda, abajo, ancho, alto.
axes0.plot(x,y,color='purple',linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
axes0.set_xlim([0,5])                        # Límites del eje horizontal.
axes0.set_ylim([-2,2])                       # Límites del eje vertical.
plt.show()
```

Alto y ancho especificado con el argumento 'figsize':

```python
fig0=plt.figure(figsize=(10,1))              # Alto y ancho especificado con figsize.
axes0 = fig0.add_axes([0,0,1,1])             # Izquierda, abajo, ancho, alto.
axes0.plot(x,y,color='red',linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
plt.show()
```

#### 1.2. Múltiples gráficos con los mismos ejes:

```python
fig0=plt.figure(figsize=(5,2), dpi=100)                      # Alto, ancho y DPI fijados.
axes0 = fig0.add_axes([0,0,1,1])                             # Izquierda, abajo, ancho, alto.
axes0.plot(x,y,color='orange',linestyle='--', label='Sin')   # 'label' para la leyenda.
axes0.plot(x,z,color='purple',linestyle='-.', label='Cos')   # 'label' para la leyenda.
axes0.legend(loc=0)                                          # Posición de la leyenda en la esquina superior derecha.
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
plt.show()
```

Salida a un fichero externo:

```python
fig0.savefig('my_fig_01.png')
```

#### 1.3. Múltiples gráficos en ejes separados:

```python
fig0=plt.figure(figsize=(8,3), dpi=80)        # Alto, ancho y DPI fijados.
axes1 = fig0.add_axes([0,0,0.3,1])            # Iquierda, abajo, ancho, alto.
axes2 = fig0.add_axes([0.5,0,0.3,1])          # Izquierda, abajo, ancho, alto.
axes1.plot(x,y,color='red',linestyle=':')  
axes2.plot(x,z,color='blue',linestyle='-.')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('SINE')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('COSINE')
plt.show()
```

```python
fig0=plt.figure(figsize=(8,4), dpi=80)           # Alto, ancho y DPI fijados.
axes1 = fig0.add_axes([0,0.5,0.2,0.5])           # Iquierda, abajo, ancho, alto.
axes2 = fig0.add_axes([0.5,0,0.2,0.5])           # Iquierda, abajo, ancho, alto.
axes1.plot(x,z,linestyle='none',marker='^',markersize=10,markerfacecolor='purple',markeredgecolor='red',alpha=0.5)
axes2.plot(x,y,color='blue',linestyle='--')  
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('COSINE')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('SINE')
plt.show()
```

#### 1.4. Múltiples gráficos en una matriz de ejes:

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
# (0,0)
c = np.random.randn(10000)
axes[0,0].hist(c,bins=100,color='green',density=True,alpha=0.3)
axes[0,0].set_title('HISTOGRAM')
# (1,0)
axes[1,0].plot(x,z,linestyle='none',marker='^',markersize=10,markerfacecolor='purple',markeredgecolor='red',alpha=0.5)
axes[1,0].set_xlabel('X')
axes[1,0].set_ylabel('Y')
axes[1,0].set_title('COSINE')
# (0,1)
a = np.array(['Q1', 'Q2', 'Q3','Q4'])
b = np.array([ 234.0, 254.7, 144.6, 317.6])
axes[0,1].bar(a,b,color='blue',alpha=0.5)
axes[0,1].set_title('BAR PLOT')
# (1,1)
axes[1,1].plot(x,y,color='green',linestyle=':')  
axes[1,1].set_xlabel('X')
axes[1,1].set_ylabel('Y')
axes[1,1].set_title('SINE')

plt.tight_layout()                             # Evitar la superposición.
plt.show()
```

Salida a un fichero externo:

```python
fig.savefig('my_fig_02.png')                           
```
