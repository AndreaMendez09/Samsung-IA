### 1. Visualizaci칩n con Seaborn (m치s ejemplos): 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')           # desactivar las alarmas. 
```

#### 1.1. Matriz de diagramas de dispersi칩n:

```python
# Leer los datos 'iris' desde la biblioteca.
dat = sns.load_dataset('iris')
```

```python
# Una matriz de diagramas de dispersi칩n.
sns.pairplot(dat)
plt.show()
```

```python
# Colorea los marcadores con 'especies'.
sns.pairplot(dat, hue='species')
plt.show()
```

```python
# Colorea los marcadores con 'especies'. Aplica una paleta.
sns.pairplot(dat, hue='species', palette='coolwarm')
plt.show()
```

#### 1.2. PairGrid:

```python
# Una matriz de diagramas de dispersi칩n.
g=sns.PairGrid(dat)
g.map(plt.scatter)
plt.show()
```

```python
# Una matriz de diferntes diagramas.
g=sns.PairGrid(dat)
g.map_diag(sns.distplot, color='orange')         # Diagonal = histograma.
g.map_upper(plt.scatter, color='green')          # Tri치ngulo superior = diagrma de dispersi칩n.
g.map_lower(sns.kdeplot, color='blue')           # Tri치ngulo inferior = KDE.
plt.show()
```

#### 1.3. Rejilla de facetas (FacetGrid):

```python
# M칰ltiples histogramas.
g=sns.FacetGrid(data=dat, col='species')
g.map(sns.distplot, 'petal_length')
plt.show()
```

```python
# M칰ltiples histogramas.
g=sns.FacetGrid(data=dat, row='species')
g.map(sns.distplot, 'petal_length')
plt.show()
```

```python
# M칰ltiples diagramas de dispersi칩n.
g=sns.FacetGrid(data=dat, col='species')
g.map(plt.scatter, 'petal_length', 'petal_width',color='orange')
plt.show()
```

#### 1.4. Mapa de calor:

```python
# Leer los datos 'mpg' desde la librer칤a. 
dat = sns.load_dataset('mpg')
```

```python
# Matriz de correlaci칩n.
x = dat.corr()
x
```

```python
# Mapa de calor por defecto.
sns.heatmap(x)
plt.show()
```

```python
# Mapa de calor.
# Utilizar 'cmap' en lugar de 'palette'.
sns.heatmap(x, cmap = 'coolwarm')
plt.show()
```

```python
# Mapa de calor.
# Ajustar el centro.
sns.heatmap(x, cmap = 'coolwarm', center=-1)
plt.show()
```
