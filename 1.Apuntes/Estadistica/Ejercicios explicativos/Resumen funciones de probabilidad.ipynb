```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Función de distribución
**Nombre en inglés:** Cumulative Distribution Function

**Abreviatura**: cdf

**Interpretación:** 

$$\textrm{cdf}(x) = \mathbb{P}(X \leq x)$$

```python
# CDF para la normal
x = np.linspace(-10,10,1000)
y = stats.norm.cdf(x)

plt.plot(x,y)
plt.show()
```

## Función de masa
**Nombre en inglés:** Probability Mass Function

**Abreviatura**: pmf

**Interpretación:** En distribuciones discretas

$$\textrm{cdf}(x) = \sum_{x_i \leq x} \textrm{pmf}(s)\,ds$$

```python
# PMF para la binomial
x = np.array(range(10 + 1))
y = stats.binom.pmf(x, n = 10, p = 0.3)

plt.scatter(x,y)
plt.show()
```

## Función de densidad
**Nombre en inglés:** Probability Density Function

**Abreviatura**: pdf

**Interpretación:** En distribuciones continuas,

$$\textrm{cdf}(x) = \int_{-\infty}^x \textrm{pdf}(s)\,ds$$

```python
# PDF para la normal
x = np.linspace(-10,10,1000)
y = stats.norm.pdf(x)

plt.plot(x,y)
plt.show()
```

## Función cuantílica
**Nombre en inglés:** Percent Point Function

**Abreviatura**: ppf

**Interpretación:** Inverso de la función de distribución

$$
    \mathbb{P}(X \leq \textrm{ppf}(\alpha)) = \alpha
$$

```python
# CDF para la normal
x = np.linspace(-10,10,1000)
y = stats.norm.ppf(x)

plt.plot(x,y)
plt.show()
```
