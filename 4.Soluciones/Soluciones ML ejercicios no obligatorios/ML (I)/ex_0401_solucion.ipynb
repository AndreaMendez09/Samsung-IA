### 1. Agrupación de K-Medias con datos simulados:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
%matplotlib inline
```

#### 1.1. Genere datos simulados y visualice:

```python
# Dataset #1.
X1, label1 = make_blobs(n_samples=100, n_features=2, centers=2, cluster_std = 5, random_state=123)
plt.scatter(X1[:,0],X1[:,1], c= label1, alpha=0.7 )
plt.title('Dataset #1 : Original')
plt.show()
```

```python
# Dataset #2
X2, label2 = make_blobs(n_samples=100, n_features=2, centers=3, cluster_std = 1, random_state=321)
plt.scatter(X2[:,0],X2[:,1], c= label2, alpha=0.7 )
plt.title('Dataset #2 : Original')
plt.show()
```

#### 1.2. Aplicar clústeres de k-medias y visualizar:

Aplique el algoritmo K-Means sobre los datasets anteriores y visualice los mismos coloreando los puntos de cada clúster con un color diferente.



```python
# Dataset #1 y dos clusters.
kmeans = KMeans(n_clusters=2,random_state=123)                 # objeto kmeans para 2 clústeres. radom_state = 123 significa inicialización determinista.
kmeans.fit(X1)                                                 # Aprendizaje no supervisado => Solo X1.  
myColors = {0:'red',1:'green', 2:'blue'}                       # Definir una paleta de colores: 0~2.
plt.scatter(X1[:,0],X1[:,1], c= pd.Series(kmeans.labels_).apply(lambda x: myColors[x]), alpha=0.7 )   
plt.title('Dataset #1 : K-Means')
plt.show()
```

```python
# Dataset #1 and three clusters.
kmeans = KMeans(n_clusters=3,random_state=123)                 # objeto kmeans para 3 clústeres. radom_state = 123 significa inicialización determinista.
kmeans.fit(X1)                                                 # Aprendizaje no supervisado => Solo X1.    
plt.scatter(X1[:,0],X1[:,1], c= pd.Series(kmeans.labels_).apply(lambda x: myColors[x]), alpha=0.7 ) 
plt.title('Dataset #1 : K-Means')
plt.show()
```

```python
# Dataset #2 and two clusters.
kmeans = KMeans(n_clusters=2,random_state=123)                 # objeto kmeans para 2 clústeres. radom_state = 123 significa inicialización determinista.
kmeans.fit(X2)                                                 # Aprendizaje no supervisado => Solo X2.    
plt.scatter(X2[:,0],X2[:,1], c= pd.Series(kmeans.labels_).apply(lambda x: myColors[x]), alpha=0.7 )  
plt.title('Dataset #2 : K-Means')
plt.show()
```

```python
# Dataset #2 and three clusters.
kmeans = KMeans(n_clusters=3, random_state=123)                # objeto kmeans para 3 clústeres. radom_state = 123 significa inicialización determinista.
kmeans.fit(X2)                                                 # Aprendizaje no supervisado => Solo X2.  
plt.scatter(X2[:,0],X2[:,1], c= pd.Series(kmeans.labels_).apply(lambda x: myColors[x]), alpha=0.7 )  
plt.title('Dataset #2 : K-Means')
plt.show()
```

NOTA: los grupos en k-medias tienen límites lineales. El resultado de la agrupación puede revelar la estructura existente, pero esto no siempre está garantizado.
