### Ejemplo 1 con KMeans

```python
#Importamos el dataset que vamos a utilizar
from sklearn.datasets import load_iris
data = load_iris()
```

```python
X = data['data'] #Es la información que tenemos de cada columna (salvo la etiqueta)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 3) #Sabemos que tenemos tres clases por el dataset, normalmente no lo tendremos
kmeans.fit(X) #Le pasamos la información
```

```python
kmeans.labels_ #Las etiquetas las pone aleatoriamente, es decir en este caso de ejecución para concordar
#Con nuestras labels le a puesto a las 0 y 1 y viceversa, pero lo que importa es la agrupacion
```

```python
kmeans.cluster_centers_
```

### Ejemplo 2 con DBSCAN

```python
Y = data['data']
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=2)
dbscan.fit(Y)
```

```python
dbscan.labels_
```

```python
dbscan = DBSCAN(eps=0.6, min_samples=5)
dbscan.fit(Y)
dbscan.labels_
```

```python
#Para intentar encontrar estos hiperparametros podemos usuar
import numpy as np
for eps in np.arange(0.1,1,0.1):
    for min_samples in range (2,20):
        dbscan2 = DBSCAN(eps=eps, min_samples=min_samples)
        dbscan2.fit(Y)
        print('------')
        print(eps)
        print(min_samples)
        print(len(set(dbscan2.labels_)))
        print(list(dbscan2.labels_).count(-1))
```
