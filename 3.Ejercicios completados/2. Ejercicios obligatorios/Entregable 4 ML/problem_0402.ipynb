study = una matriz que contiene las horas de estudio. Esta es la variable explicativa. <br>
score = una matriz que contiene las puntuaciones de las pruebas. Esta es la variable de respuesta.

```python
study = np.array([ 3, 4.5, 6, 1.2, 2, 6.9, 6.7, 5.5]).reshape(-1,1)
score = np.array([ 88, 85, 90, 80, 81, 92, 95, 90]).reshape(-1,1)
```

1). Defina la clase 'linerRegression' para que coincida con el resultado que se muestra a continuación:

```python
class linearRegression:
    def __init__(self):
        self.r2 = 0
        self.m = 0
        self.c = 0
        
    def train(self, input_X, input_Y):
        #Primero inicializo todas mis variables
        mean_x = np.mean(study)
        mean_y = np.mean(score)
        self.m = len(study)
        numer = 0
        denom = 0
        numer2 = 0 
        denom2 = 0
        
        # Usamos la formula para calcular m
        for i in range(self.m):
            numer += (study[i] - mean_x) * (score[i] - mean_y) #Es la parte del numerador
            denom += (study[i] - mean_x) ** 2 #Es la parte del denominador

        self.m = numer / denom #Actualizamos m
        self.c = mean_y - (self.m * mean_x) #Teniendo todos los datos salvo la c, despejamos la c y calculamos
        
        # Calculamos la R^2, que mide lo "bueno" que es el modelo
        for i in range(len(study)):
            y_pred = self.c + self.m * study[i] # Utilizamos la formula vista anteriormente para predecir y
            numer2 += (score[i] - y_pred) ** 2 #Es la parte del numerador
            denom2 += (score[i] - mean_y) ** 2 #Es la parte del denominador

        self.r2 = 1 - (numer2/denom2)
    
    def output(self):
        print(f'Intercept   {self.c}\nX1          {self.m}\n------------------------\nR^2: {self.r2}')
```

**Ejecución:**

```python
LR = linearRegression()                    # Crea un objeto de regresión lineal.
LR.train(study, score)                     # Entrenar.
LR.output()                                # Muestra el resultado del entrenamiento.
```
