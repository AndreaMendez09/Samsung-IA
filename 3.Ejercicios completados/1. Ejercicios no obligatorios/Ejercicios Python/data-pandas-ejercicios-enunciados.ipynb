# Ejercicios de introducción a Pandas (enunciados)

Dado que muchos usuarios potenciales de pandas tienen cierta familiaridad con `SQL`, estos ejercicios pretenden proporcionar algunos ejemplos de cómo se realizarían varias operaciones de SQL usando pandas.

````python
import pandas as pd
pd.__version__
import os
df = pd.read_csv('tips.csv', header='infer')
````

**Paso inicial**: cargar el dataset **tips** (`'../data/teoria/tips.csv'`) y mostrar sus 5 primeras filas:
![tips head](../images/tips.png)

````python
df.head(5)
````

```sql
SELECT total_bill, tip, smoker, time
FROM tips
LIMIT 7;
```

````python
df[['total_bill', 'tip', 'smoker', 'time']].head(7)
````

```sql
SELECT *
FROM tips
WHERE time = 'Dinner'
LIMIT 5;
```

````python
df.where(df['time'] == 'Dinner').head(5)
````

```sql
SELECT count(*)
FROM tips
WHERE time = 'Dinner';
```

````python
df[df['time'] == 'Dinner'].count() #Preguntar al profe
````

```sql
SELECT *
FROM tips
WHERE time = 'Dinner' AND tip > 5.00;
```

````python
df[(df['time'] == 'Dinner') & (df.tip > 5.00)]
````

```sql
SELECT sex, count(*)
FROM tips
GROUP BY sex;
```

````python
df[['sex']].value_counts()
````

```sql
SELECT day, AVG(tip), COUNT(tip)
FROM tips
GROUP BY day;
```

```sql
SELECT day, AVG(tip), COUNT(tip), MAX(total_bill), MIN(total_bill)
FROM tips
GROUP BY day;
```

```sql
SELECT smoker, day, COUNT(tip), AVG(tip)
FROM tips
GROUP BY smoker, day;
```

Sean los siguientes DataFrame:

````python
from numpy.random import randn 
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': randn(4)})

df2 = pd.DataFrame({'key': ['B', 'D', 'D', 'E'],
                    'value': randn(4)})
````

```sql
SELECT *
FROM df1
INNER JOIN df2
  ON df1.key = df2.key;
```

```sql
SELECT *
FROM df1
LEFT OUTER JOIN df2
  ON df1.key = df2.key;
```

```sql
SELECT *
FROM df1
RIGHT OUTER JOIN df2
  ON df1.key = df2.key;
```

```sql
SELECT *
FROM df1
FULL OUTER JOIN df2
  ON df1.key = df2.key;
```

Sean los siguientes DataFrame:

````python
df1 = pd.DataFrame({'city': ['Chicago', 'San Francisco', 'New York City'],
                    'rank': range(1, 4)})

df2 = pd.DataFrame({'city': ['Chicago', 'Boston', 'Los Angeles'],
                    'rank': [1, 4, 5]})
````

```sql
SELECT city, rank
FROM df1
UNION ALL
SELECT city, rank
FROM df2;
```

```sql
SELECT city, rank
FROM df1
UNION
SELECT city, rank
FROM df2;
```

```sql
UPDATE tips
SET tip = tip*2
WHERE tip < 2;
```

```sql
DELETE FROM tips
WHERE tip > 9;
```
