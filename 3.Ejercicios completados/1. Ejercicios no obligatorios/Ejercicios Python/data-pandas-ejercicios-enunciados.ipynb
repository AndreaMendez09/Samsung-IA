Sean los siguientes DataFrame:

````python
from numpy.random import randn 
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': randn(4)})

df2 = pd.DataFrame({'key': ['B', 'D', 'D', 'E'],
                    'value': randn(4)})
````

```sql
SELECT *
FROM df1
INNER JOIN df2
  ON df1.key = df2.key;
```

```sql
SELECT *
FROM df1
LEFT OUTER JOIN df2
  ON df1.key = df2.key;
```

```sql
SELECT *
FROM df1
RIGHT OUTER JOIN df2
  ON df1.key = df2.key;
```

```sql
SELECT *
FROM df1
FULL OUTER JOIN df2
  ON df1.key = df2.key;
```

Sean los siguientes DataFrame:

````python
df1 = pd.DataFrame({'city': ['Chicago', 'San Francisco', 'New York City'],
                    'rank': range(1, 4)})

df2 = pd.DataFrame({'city': ['Chicago', 'Boston', 'Los Angeles'],
                    'rank': [1, 4, 5]})
````

```sql
SELECT city, rank
FROM df1
UNION ALL
SELECT city, rank
FROM df2;
```

```sql
SELECT city, rank
FROM df1
UNION
SELECT city, rank
FROM df2;
```
