### 2. Expresiones regulares:

```python
import re 
```

#### 2.1. Metacaracter: . 

```python
# Cadenas de prueba.
x1 = "ct"
x2 = "cat"
x3 = "caat"
```

```python
# Una expresión regular.
my_expression = re.compile("c.t")
```

```python
if my_expression.search(x1):
    print("Yes!")
else:
    print("No")
```

```python
if my_expression.search(x2):
    print("Yes!")
else:
    print("No")
```

```python
if my_expression.search(x3):
    print("Yes!")
else:
    print("No")
```

#### 2.2. Metacaracter: ?

```python
# Una expresión regular.
my_exp = "ca?t"
```

```python
if re.search(my_exp, x1):
    print("Yes!")
else:
    print("No")
```

```python
if re.search(my_exp, x2):
    print("Yes!")
else:
    print("No")
```

```python
if re.search(my_exp, x3):
    print("Yes!")
else:
    print("No")
```

#### 2.3. Metacaracter: {m,n}

```python
# Cadenas de prueba.
x1 = "ct"
x2 = "cat"
x3 = "caat"
x4 = "caaat"
```

```python
# Una expresión regular.
my_exp= "ca{1,2}t"
```

```python
if re.search(my_exp, x1):
    print("Yes!")
else:
    print("No")
```

```python
if re.search(my_exp, x2):
    print("Yes!")
else:
    print("No")
```

```python
if re.search(my_exp, x3):
    print("Yes!")
else:
    print("No")
```

```python
if re.search(my_exp, x4):
    print("Yes!")
else:
    print("No")
```

#### 2.4. Metacaracter: ^

```python
# Cadenas de prueba.
x1 = "That is your book."
x2 = "This is my book."
```

```python
# Una expresión regular.
my_exp = "^This"
```

```python
if re.search(my_exp, x1):
    print("Yes!")
else:
    print("No")
```

```python
if re.search(my_exp, x2):
    print("Yes!")
else:
    print("No")
```

#### 2.5. Metacaracter: ^

```python
# Cadenas de prueba.
x1 = "I don't like you"
x2 = "You are great"
x3 = "in case you forget, remember this"
```

```python
# Una expresión regular.
my_exp = "^[Ii]"
```

```python
if re.search(my_exp, x1):
    print("Yes!")
else:
    print("No")
```

```python
if re.search(my_exp, x2):
    print("Yes!")
else:
    print("No")
```

```python
if re.search(my_exp, x3):
    print("Yes!")
else:
    print("No")
```
